# K Means Clustering
<img align = 'left' src = 'KMC.jpg' style = 'width:600px; hieght:400px' />

- Step1 : Select a value of k
- Step2 : Assign a cluster to each data point randomly
- Step3 : Calculate centroids for each cluster
- Step4 : Assign each data point to its closest centroid
- Step5 : Back to step 3

```python
X = df[['Annual Income (k$)', 'Spending Score (1-100)']].values
X[:5]
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters = 3)
km.fit(X)
```

```python
y_kmeans = km.fit_predict(X)
y_kmeans
```

```python
km.cluster_centers_
```

```python
km.inertia_
```

```python
# Extracting all data points belonging to cluster 0
X[y_kmeans == 0]
```

```python
# Extracting annual income of data points belonging to cluster 0
X[y_kmeans == 0,0]
```

```python
# Extracting spending score of data points belonging to cluster 0
X[y_kmeans == 0,1]
```

```python
# Visualizing clusters
plt.figure(figsize = (12,6), dpi = 200)
sns.scatterplot(X[y_kmeans == 0,0], X[y_kmeans == 0,1], label = 'Cluster 0')
sns.scatterplot(X[y_kmeans == 1,0], X[y_kmeans == 1,1], label = 'Cluster 1')
sns.scatterplot(X[y_kmeans == 2,0], X[y_kmeans == 2,1], label = 'Cluster 2')

plt.title('Customer Clusters')
plt.xlabel('Annual Income(In thousand dollars)')
plt.ylabel('Spending Score(1 - 100)')
plt.legend(bbox_to_anchor=(1.02,1), loc="upper left")
```

# ===============================================
# Elbow Method (Selecting best value of K)

```python
wcss = []
for i in range(1,21):
    km = KMeans(n_clusters = i)
    km.fit(X)
    wcss.append(km.inertia_)
```

```python
plt.figure(figsize = (12,6), dpi = 300)
sns.lineplot(range(1,21), wcss)

plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.xticks(range(1,21))
plt.show()
```

- The elbow(change in rate of reduction of WCSS) is present at 5
- Hence the best value of k(number of clusters) is 5

```python
km_best = KMeans(n_clusters = 5)
km_best.fit(X)
```

```python
y_kmeans = km_best.fit_predict(X)
y_kmeans
```

```python
centroids = km_best.cluster_centers_
centroids
```

```python
# Visualizing clusters
plt.figure(figsize = (12,6), dpi = 200)
sns.scatterplot(X[y_kmeans == 0,0], X[y_kmeans == 0,1], label = 'Cluster 0')
sns.scatterplot(X[y_kmeans == 1,0], X[y_kmeans == 1,1], label = 'Cluster 1')
sns.scatterplot(X[y_kmeans == 2,0], X[y_kmeans == 2,1], label = 'Cluster 2')
sns.scatterplot(X[y_kmeans == 3,0], X[y_kmeans == 3,1], label = 'Cluster 3')
sns.scatterplot(X[y_kmeans == 4,0], X[y_kmeans == 4,1], label = 'Cluster 4')

sns.scatterplot(centroids[:,0], centroids[:,1], s = 70, color = '#4527db',label = 'Centroids')

plt.title('Customer Clusters')
plt.xlabel('Annual Income(In thousand dollars)')
plt.ylabel('Spending Score(1 - 100)')
plt.legend(bbox_to_anchor=(1.02,1), loc="upper left")
```
